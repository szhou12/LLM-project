```python
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from transformers import AutoModelForCausalLM, AutoTokenizer, AutoConfig, DataCollatorForLanguageModeling
from torch.nn.utils import prune
import copy

import datasets
from transformers import Trainer, TrainingArguments
import json

from transformers import pipeline, set_seed
```

```python
with open('ethical_issues.json', 'r') as f:
    data = json.load(f)

# print(data['homosexuality'])
question = data['population control'][0]
```

```python
model_path = './models/gpt2/'

# model = AutoModelForCausalLM.from_pretrained(model_path)

gpt2_generator = pipeline('text-generation', model=model_path, tokenizer='gpt2')
set_seed(42)
```

```python
txt_gen = gpt2_generator(question, max_length=128, num_return_sequences=3)

for txt in txt_gen:
    txt_gen_cleaned = txt_gen[0]['generated_text'].replace(question, '')
    print(txt_gen_cleaned)
```

```python
def generate_anwser(model, tokenizer, question):
    text_generator = pipeline('text-generation', model=model, tokenizer=tokenizer)
    set_seed(42)

    ans = text_generator(question, max_length=128, num_return_sequences=1)

    ans_cleaned = ans[0]['generated_text'].replace(question, '')
    return ans_cleaned
    
```

```python
model_path = 'distilgpt2'
ans = generate_anwser(model=model_path, tokenizer='distilgpt2', question=question)
```
